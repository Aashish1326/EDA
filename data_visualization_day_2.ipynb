##### import the titanic dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
data=pd.read_csv(r"D:\Innomatics\Data_visualization\Data_visualization_day_2\titanic.csv")
data.head()
```

##### 1. Write a Python program to display a bar chart of the fare of class.

```python
### code here
x=data[data['class']=='First']['fare'].mean()
y=data[data['class']=='Second']['fare'].mean()
z=data[data['class']=='Third']['fare'].mean()
plt.bar(['First','Second','Third'],[x,y,z])
plt.ylabel('avg of fare')
plt.show()
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
### code here
plt.barh(['First','Second','Third'],[x,y,z])
plt.xlabel('avg of fare')
plt.show()
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
### code here
# method 1
x=data[data['class']=='First']['fare'].mean()
y=data[data['class']=='Second']['fare'].mean()
z=data[data['class']=='Third']['fare'].mean()
c = ["orange", "green","blue",]

plt.bar(['First','Second','Third'],[x,y,z],color=c)
plt.ylabel('avg of fare')
plt.show()
```

```python
#method 2

x=data[data['class']=='First']['fare'].mean()
y=data[data['class']=='Second']['fare'].mean()
z=data[data['class']=='Third']['fare'].mean()

sns.barplot(['First','Second','Third'],[x,y,z])
plt.ylabel('avg of fare')
plt.xticks(rotation=90)
plt.show()
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
### code here
plt.pie([x,y,z],labels=["First","Second","Third"],autopct="%0.2f%%",explode=[0,0.1,0],radius=2)
plt.title("fare of final marks")
plt.show()
```
